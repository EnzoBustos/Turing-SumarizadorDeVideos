# Import libraries

```python
import yt_dlp
import whisper
import os
from transformers import pipeline
import torch
from summarizer import Summarizer
import datetime
import time
```

# Download audio from youtube video

```python
st = datetime.datetime.now()

print("Starting downloading part!")

def download_audio(link):
  with yt_dlp.YoutubeDL({'extract_audio': True, 'format': 'bestaudio', 'outtmpl': '%(title)s.mp3'}) as video:
    info_dict = video.extract_info(link, download = True)
    video_title = info_dict['title']
    video.download(link)
    return video_title

video_title = download_audio('https://www.youtube.com/watch?v=n3Xv_g3g-mA')
#video_title = download_audio('https://www.youtube.com/watch?v=sNhhvQGsMEc')

et = datetime.datetime.now()

print('Execution time for downloading:', et - st, 'seconds')
```

# speech-to-text

```python
st = datetime.datetime.now()

print("Starting speech-to-text part!")

model = whisper.load_model("base.en")
directory = os.getcwd()
result = model.transcribe(os.path.join(directory, video_title + ".mp3"))
#result = model.transcribe("/content/The Fermi Paradox — Where Are All The Aliens？ (1⧸2).mp3")
transcript_text = result["text"]

et = datetime.datetime.now()

print('Execution time for speech-to-text:', et - st, 'seconds')
```

# Saving transcript

```python
with open(video_title + ".txt", "w") as text_file:
    text_file.write(transcript_text)
'''
with open("Fermi Paradox" + ".txt", "w") as text_file:
    text_file.write(transcript_text)
'''
```

# Extrative part

```python
st = datetime.datetime.now()

print("Starting extractive part!")

model = Summarizer()
extractive = model(transcript_text, num_sentences=15)

et = datetime.datetime.now()

print('Execution time for extractive:', et - st, 'seconds')
```

```python
extractive
```

# Abstractive part

```python
st = datetime.datetime.now()

print("Starting abstractive part!")

device_num = 0 if torch.cuda.is_available() else -1
model_name = "facebook/bart-large-cnn"
summarizer = pipeline("summarization", model=model_name, tokenizer=model_name, device=device_num)
output_text = summarizer(extractive)[0]["summary_text"]

et = datetime.datetime.now()

print('Execution time for abstractive:', et - st, 'seconds')
```
